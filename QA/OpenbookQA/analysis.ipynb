```python
import pandas as pd
```

```python
data = pd.read_csv("result/mc_gpt_4.csv",index_col=0)
data.head(5)
```

```python

from rouge_score import rouge_scorer
scorer = rouge_scorer.RougeScorer(['rouge1','rouge2','rougeL'], use_stemmer=True)
```

```python
f_score = 0
acc = 0
slot_ans_list = []
answer_list = []
for index,row in data.iterrows():
    answer_key = row['Answer Key']
    if answer_key != 'D':
        answer = row['Complete Question'].split(answer_key+") ")[1].split(" (")[0]
    else:
        answer = row['Complete Question'].split(answer_key+") ")[1]
    # print(row['slot_answer'])
    if (")") in row['slot_answer']:
        slot_answer = row['slot_answer'].split(")")[1]
        if slot_answer[0] == " ":
            slot_answer = slot_answer[1:]
    rouge_score_tmp = scorer.score(slot_answer,answer)
    f_score_tmp = rouge_score_tmp['rougeL'].fmeasure
    f_score += f_score_tmp
    print(answer,"|",slot_answer)
    slot_ans_list.append(slot_answer)
    answer_list.append(answer)
    if slot_answer == answer:
        acc += 1
print("Acc: ",acc*1.0/data.shape[0])
print("Rouge-l F1: ",f_score*1.0/data.shape[0])
    
```

```python
data['slot_answer'] = slot_ans_list
data['answer'] = answer_list
data.to_csv("result/mc_gpt4.csv")
```
